```python
import pandas as pd
import numpy as np

# Read the dataset
data = pd.read_excel('/content/Dry_Bean_Dataset.xlsx')
data.head()
```

```python
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler

#scaler = MinMaxScaler()
#df_normalized = scaler.fit_transform(data)

scaler = MinMaxScaler()
normalized_data= scaler.fit_transform(data.iloc[:, :-1])
print(normalized_data)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.neighbors import NearestNeighbors

i = 1234
j = 7

# Backup the original value for printing
original_value = normalized_data[i, j]

# Find k nearest neighbors
nn = NearestNeighbors(n_neighbors=5)
nn.fit(normalized_data)
distances, indices = nn.kneighbors(normalized_data)
nearest_neighbors = normalized_data[indices[i], :]

# Filter neighbors by the same class as the target sample
target_class = data.iloc[i, -1]
same_class_indices = indices[i][data.iloc[indices[i], -1] == target_class]
same_class_neighbors = normalized_data[same_class_indices, :]

# Calculate simple average
simple_average = np.mean(same_class_neighbors[:, j])

# Calculate weighted average
ep = 1e-5
k_used = min(5, len(same_class_indices))  # Ensure k is not greater than available neighbors
weights = 1.0 / (distances[i, :k_used]+ ep)
weighted_average = np.sum(same_class_neighbors[:, j] * weights) / np.sum(weights)

# Print results
print(f"(i = {i}, j = {j})")
print(f"Original: {original_value:.6f}")
print(f"S_A: {simple_average:.6f}")
print(f"W_A: {weighted_average:.6f}")
```
